Series 타입
- 동일한 성격의 데이터를 나열한것(열 한나만 존재)
- 데이터형: 문자열, 실수, 정수, 카테고리, 시간 등

```python
import pandas as pd

data = pd.Series(['서울', '부산', '대전', '대구', '광주'])  # 리스트/배열/튜블
data
```

```python
# 딕셔너리: 키 => index 적
data = pd.Series({1:'서울', 2:'부산', 3:'대전', 4:'대구', 5:'광주'})
data
```

#### DataFrame
- 데이터프레임(DataFrame)은 테이블 형태로 구성된 2차원 데이터 구조입니다.
- 행(row)과 열(column)로 이루어져 있으며, 각 열은 서로 다른 데이터 타입을 가질 수 있습니다.

```python
# list를 이용한 DataFrame 생성(list는 2차원 구조)
import pandas as pd

tmp = [[56, 92, 88], [78, 86, 91], [92, 33, 89]]
df = pd.DataFrame(tmp, columns=['국어', '영어', '수학'])
df
```

```python
# dict를 이용한 DataFrame 생성
import pandas as pd

tmp = {'이름':['홍길동', '김생성', '박이용'],
       '국어':[86, 92, 71],
       '영어':[88, 76, 72],
       '수학':[82, 73, 93]
       }

df = pd.DataFrame(tmp)  # dict의 Key가 열 Index(이름) 적용
df
```

DataFrame에서 데이터 조회

```python
# 열단위 데이터 가져오기
print(df['이름'])   # 이름 열 출력
print(df.이름)      # 이름 열 출력
```

```python
print(df[['이름', '수학']])   # 이름/수학 열 출력(2개 이상의 열값 출력)
display(df[['이름', '수학']])  # display(): 데이터프레임형 데이터를 서식을 적용해 출력
```

#### 행단위 데이터 조회(키index, 데이터index)
- index 값
    - df.loc[ 인덱스 이름 ]
    - df.loc[ [인덱스이름1, 인덱스이름2] ]
- index 위치(데이터 위치 index)
    - df.iloc[인덱스 번호]
    - df.iloc[ 시작 : 끝 ]


```python
df
```

```python
# 행/열 단위 데이터 조회(index값)
df.loc[1, : ]
```

```python
df.loc[1, '국어']
```

```python
display(df.loc[1:2, ['국어','수학']])
display(df.loc[:, ['국어','수학']])
```

```python
# 행/열 단위 데이터 조회(index 위치)
display(df.iloc[0])
display(df.iloc[0, 0])
display(df.iloc[0:2, 0:3])
display(df.iloc[0:2, [0, 3, 2]])
```

#### 행/열 index 수정하기

```python
# 열index = 열이름(columns)
print(df.columns)

df.columns=['name', 'kor', 'eng', 'math']
df
```

#### 행 데이터 추가/삭제

```python
# 마지막 행에 데이터 추가
df.loc[3]=['이순신', 95, 85, 92]
df
```
