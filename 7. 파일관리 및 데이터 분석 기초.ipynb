opi_oil 폴더의 모든 데이터 병합하기

```python
import pandas as pd

df1=pd.read_excel(fdir+'opi_oil/지역_위치별(주유소).xls', header=2)
df2=pd.read_excel(fdir+'opi_oil/지역_위치별(주유소) (1).xls', header=2)
df3=pd.read_excel(fdir+'opi_oil/지역_위치별(주유소) (2).xls', header=2)
df4=pd.read_excel(fdir+'opi_oil/지역_위치별(주유소) (3).xls', header=2)
df5=pd.read_excel(fdir+'opi_oil/지역_위치별(주유소) (4).xls', header=2)

df0 = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

# 변수(객체) 삭제
del df1, df2, df3, df4, df5

df0.info()
```

- for문을 이용한 파일 병합

```python
# 병합할 대상 파일 정보 가져오기
import glob   # glob.glog(): 파일 확장자를 이용해 원하는 유형의 파일 정보만 가져오기

fnames = glob.glob(fdir + 'opi_oil/*.xls')
fnames  # 리스트 형식으로 반환
```

```python
import os     # os.listdir(): 지정한 폴더의 하위폴더와 파일명 리스트를 반환

fnames = os.listdir(fdir+'opi_oil/')
fnames
```

```python
# 파일 병합하기

df0 = pd.DataFrame() # 데이터를 누적할 DataFrame 객체 생성

for fname in fnames:
    # print(fdir + 'opi_oil/' + fname)
    df1 = pd.read_excel(fdir + 'opi_oil/' + fname, header=2)
    df0 = pd.concat([df0, df1], ignore_index=True)

del df1
df0.info()
```

```python
# 병합된 파일 저장
# index 데이터 불포함, 한글처리:cp949 => 엑셀에서 사용가능
df0.to_csv(fdir+'유가정보_230803.csv', index=False, encoding='cp949')
```

- AND / OR : 논리연산자(Boolean) => True/False
- & / | : 비트(비교)연산자(Bitwise) => 1 / 0

```python
# 셀프여부: Y, 상표:SK에너지 인 데이터 추출(두개 이상의 조건 만족: AND)
# 두개 이상의 조건: And => (조건1)&(조건2) / or => (조건1)|(조건2)
df0[(df0['셀프여부'] == "Y") & (df0['상표'] == 'SK에너지')]
```

```python
# 상표가 S-OIL 또는 SK에너지 인 데이터 추출
df0[(df0['상표']=='S-OIL')|(df0['상표']=='SK에너지')]
```
