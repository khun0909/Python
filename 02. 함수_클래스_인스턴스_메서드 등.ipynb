```python
# python 함수 정의: 특정 행동만 진행하는 함수
# 두개의 데이터를 가져와 비교하는 함수

def prnMax(a, b):
    if a > b:
        print(f'{a}가 {b}보다 큰 값입니다.')
    elif a == b:
        print(f'{a}와 {b}는 동일한 값입니다.')
    else:
        print(f'{b}가 {a}보다 큰 값입니다.')
```

```python
# 함수 호출
x=11; y=10
prnMax(x, y)
```

```python
# python 함수정의: 변수를 이용해 계산후 결과를 돌려주는 함수
def add(a, b=0):
    global aa # 전역변수
    aa = 20

    tot = a + b   # tot, pMax: 지역변수
    if a > b:
        pMax = a
    elif b > a:
        pMax = b
    else:
        pMax = "같다"

    return tot, pMax
```

```python
aa = 10; bb = 20

print(add(aa))
print('a=',aa)
```

```python
t, m =add(aa, bb)
print(t, m)
```

```python
# 매개변수 적용 함수
def user_info(name, age):
    print('이름 : %s, 나이 : %d'%(name, age))

user_info('홍길동', 35)
```

```python
# 여러개의 매개변수 입력(*변수명)

def user_info(*items):
    for item in items:
        print(item)

user_info("홍길동", 35, '경기도', '남성')
```

```python
# 일반매개변수와 *매개변수 같이 사용
def user_info(name, *items):
    print(f'name: {name}')
    for item in items:
        print(f'items: {item}')

user_info("홍길동", 35, '경기도', '남성')
```

```python
# **매개변수(**kwargs): 입력시 변수명과 값을 같이 입력받아 딕셔너리 구조로 전달
def user_info(**kwargs):
    print(kwargs)

user_info(name='홍길동', age=35, addr='경기도', height=175)
```

```python
def user_info(name, **kwargs):
    print(f'name: {name}')
    print(f'kwargs: {kwargs}')

user_info(name='홍길동', age=35, addr='경기도', height=175)
```

```python
class 빵틀:
    def 빵반죽(self):  # 메서드(멤버함수)
        self.기본재료='밀가루'   # 애트리뷰트(속성) 정의
        self.필수추가='이스트'   # 애트리뷰트(속성) 정의
        self.기본시간=5          # 애트리뷰트(속성) 정의

    def 빵모양(self):  # 메서드(멤버함수)
        self.크기가로=20   # 애트리뷰트(속성) 정의
        self.크기세로=5   # 애트리뷰트(속성) 정의
```

```python
# 빵틀 클래스를 이용한 객체(인스턴스) 생성
밤빵 = 빵틀()   # 객체(인스턴스) 생성
밤빵.추가재료 = '밤' # 클래스에 없는 객체 애트리뷰트(속성) 생성 및 값 설정
밤빵.칼로리 = 100    # 클래스에 없는 객체 애트리뷰트(속성) 생성 및 값 설정

버터빵 = 빵틀()   # 객체(인스턴스) 생성
버터빵.추가재료 = '버터'  # 클래스에 없는 객체 애트리뷰트(속성) 생성 및 값 설정
버터빵.칼로리 = 200       # 클래스에 없는 객체 애트리뷰트(속성) 생성 및 값 설정
```

```python
밤빵.빵반죽() ; 밤빵.빵모양() # 메서드(멤버함수)호출
print(f'{밤빵.기본재료}/{밤빵.필수추가}/{밤빵.기본시간}/{밤빵.추가재료}/{밤빵.칼로리}/{밤빵.크기가로}/{밤빵.크기세로}')
```

```python
## 생성자 사용예 1

class Person:
    def __init__(self):     # 생성자
        print(self, 'is generated')   # 이코드의 행번호 11, 16번의 p1,p2 객체(인스턴스) 생성명령이 실행될때 print구문이 동작하게됨
        print('Person 클래스를 실행합니다')
        self.name = 'Kate'   # 생성자(클래스메서드(멤버함수)) 애트리뷰트(속성)값  (객체멤버라고하기도함)
        self.age = 10        # 생성자(클래스메서드(멤버함수)) 애트리뷰트(속성)값  (객체멤버라고하기도함)

p1 = Person()  # 객체(인스턴스) 생성
print(p1.name, '/', p1.age)  # __init__ 생성자메서드의 애트리뷰트값을 출력

print('\n\n' + '-'*100)
p2 = Person()  # 객체(인스턴스) 생성

p2.name = 'aaron' #생성된 객체(인스턴스) p2의 name 애트리뷰트값을 설정함 (생성자에 있는 name값을 변경함.)
p2.age = 20       #생성된 객체(인스턴스) p2의 age 애트리뷰트값을 설정함 (생성자에 있는 age값을 변경함.)

print(p2.name, '/', p2.age)
```

```python
## 생성자 사용예 2

class Person:
    # 생성자에 매개변수를 선언한경우임  (self, name, age 를 매개변수라함)
    #이코드 8,9,10번행의 p1,p2,p3 객체(인스턴스) 생성시 name은 값이 있어야하고, age는 없으면 기본으로 10을 할당함
    def __init__(self, name, age=10):
        self.name = name   # 이코드 8,9,10번행의 p1,p2,p3 객체(인스턴스) 생성할때 넣어준 name값 받아옴
        self.age = age     # 이코드 8,9,10번행의 p1,p2,p3 객체(인스턴스) 생성할때 넣어준 age값 받아옴

p1 = Person('Bob', 30)  # 객체(인스턴스)생성시 생성자의 name 애트리뷰트(속성)에 'Bob', age 애트리뷰트에 30 값 argument(인자)전달
p2 = Person('Kate', 20) # 객체(인스턴스)생성시 생성자의 name 애트리뷰트(속성)에 'Kate', age 애트리뷰트에 20 값 argument(인자)전달
p3 = Person('Aaron')    # 객체(인스턴스)생성시 생성자의 name 애트리뷰트(속성)에 'Aaron' 값 argument(인자)전달
                        # age 매개변수값에 필요한 argument(인자)값을 넣지 않아서 __init__( age=10)값이 자동 할당됨

print(p1.name, p1.age)
print(p2.name, p2.age)
print(p3.name, p3.age)
```

```python
## 참고: *argument 를 활용해봅니다. (클래스내의 모든 멤버함수(메서드)에 사용가능함)
## *argument는 매개변수의 갯수가 정해져 있지 않을때 사용하는 방법입니다.
## 통상적으로 *arg라고 표현하나 *뒤의 이름은 사용자 정의어입니다.

class Person:
    def __init__(self,*data):
        self.name = data[0]
        self.age = data[1]

p1=Person('bu',30)  # pa=Person('bu',30,20,10) argument 갯수값이 많아도 __init__ 에서 data[0]과 data[1]만 사용함
p1.name, p1.age
```

```python
## 참고: **kwargs(keyword argument) 를 활용해봅니다. (클래스내의 모든 멤버함수(메서드)에 사용가능함)
## kwargs는 매개변수를 딕셔너리 형태로 처리할수 있습니다.

class Person:
    def __init__(self,**data):
        self.name = data['name']
        self.age = data['age']

p1=Person(name=['bu','song'],age=[30,50])
print(p1.name, p1.age)
print(p1.name[0], p1.age[0])
print(p1.name[1], p1.age[1])
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def sleep(self):
        print(self.name, '은 잠을 잡니다.')

a = Person('Aaron', 20)
b = Person('Bob', 30)

a.sleep()
b.sleep()
```
