```python
#########################################
####### 미션 if함수와 for
####### 점수중에 90이상의 자료만 합과 평균을 출력
#########################################
a=[70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
total=0
cnt=0

for i in a:
    if i >= 90: 
        total+=i
        cnt+=1

print(f"90점 이상\n합계점수={total}\n평균점수={total/cnt}")
# \n : 강제 개행 이스케이프 문자
```

```python
#########################################
####### 미션: 이중for
####### data[0], data[1], data[2]의 총합 출력
#########################################
data=[[80,40,14,15,45,47,45,45,78,20],
[70, 60, 55, 75, 95, 90, 80, 80, 85, 100],
[20,28,40,70,70,20,20,70,70,80]]

total=0; cnt=0

for subdata in data:
    cnt+=1
    subtotal=0
    for nm in subdata:
        subtotal+=nm

    print(f"{cnt}번째 데이터 합계: {subtotal}")
    total=total+subtotal

print(f'전체 데이터 합계:{total}')
```

```python
data=[[80,40,14,15,45,47,45,45,78,20],
      [70, 60, 55, 75, 95, 90, 80, 80, 85, 100],
      [20,28,40,70,70,20,20,70,70,80]]

total_lst=[]

for subdata in data:
    subtotal=0
    for nm in subdata:
        subtotal+=nm
    total_lst.append(subtotal)

print("▶ 첫번째 자료는 ===> {0}\n   합계 ==> {1}".format(data[0], total_lst[0]))
print("▶ 두번째 자료는 ===> {0}\n   합계 ==> {1}".format(data[1], total_lst[1]))
print("▶ 세번째 자료는 ===> {0}\n   합계 ==> {1}".format(data[2], total_lst[2]))
print(f"▶ 전체합계 ===> {sum(total_lst)}")
```

```python
#########################################
####### 미션: 이중for
####### data[0], data[1], data[2]의 각각의 최대값/최소값/합계 출력
####### 전체 최대값/최소값/합계 출력
#########################################
data=[[80,40,14,15,45,47,45,45,78,20],
      [70, 60, 55, 75, 95, 90, 80, 80, 85, 100],
      [20,28,40,70,70,20,20,70,70,80]]

sublist=[]
total=0;tmax=data[0][0];tmin=data[0][0]

for subdata in data:
    submax=subdata[0];submin=subdata[0];subtot=0
    for nm in subdata:
        subtot=subtot+nm
        if nm > submax:
            submax=nm
        if nm < submin :
            submin=nm

    sublist.append([submax, submin, subtot])
    total+=subtot
    if submax > tmax: tmax=submax
    if submin < tmin: tmin=submin

cnt=0
for prn in sublist:
    print(f"▶ {cnt+1}번째 자료 ===> {data[cnt]}\n   최대값 ==> {prn[0]}, 최소값 ==> {prn[1]}, 합계 ==> {prn[2]}\n")
    cnt=+1
   
print("▶ 전체 데이터")
print(f"   최대값 ==> {tmax}, 최소값 ==> {tmin}, 전체합계 ==> {total}")
```

```python
data=[[80,40,14,15,45,47,45,45,78,20],
      [70, 60, 55, 75, 95, 90, 80, 80, 85, 100],
      [20,28,40,70,70,20,20,70,70,80]]

sublist=[]
total=[]

for d in data:
    sublist.append([max(d), min(d), sum(d), sum(d)/len(d)])
    total = total + d

print(f"처리결과(sublist) ==> {sublist}")
print("전체 최대값 ==> %d / 전체 최소값 ==> %d / 전체 합계 ==> %d, 전체 평균 ==> %.2f"%(max(total), min(total), sum(total), sum(total)/len(total)))
```

```python
#########################################
####### 미션
######  구글검색, 파이썬 문자 split
#########################################

# 아래의 문자열을 txt append에 '/' 단위로 구별하여서 추가
# print해서 확인 txt[0]=> 노트북,  txt[1]=>모니터,  txt[2]=>키보드가 됨
pumMok='노트북/모니터/키보드'

txt=[];tmp=""

for t in pumMok:
    if t == "/":
        txt.append(tmp)
        tmp=""
    else:
        tmp = tmp + t

if t != "/": 
    txt.append(tmp)

print(txt)
```

```python
#########################################
####### 미션
######  for를 이용하여 문자열을 나누고 txt에 append
#########################################

# '/'를 기준으로 나눔
# txt 의 최종결과는
# 노트북,모니터,키보드,키보드,볼펜,사인펜,노트,usb,형광펜

pumMok=['노트북/모니터/키보드',
        '키보드/볼펜',
        '사인펜/노트/usb/형광펜']

txt=[]

for pm in pumMok:
    tmp=""
    for t in pm:
        if t == "/":
            txt.append(tmp)
            tmp=""
        else:
            tmp += t

    if t != "/": 
        txt.append(tmp)

print(txt)

   
```

```python
#########################################
####### 미션
######  for를 이용하여 문자열을 나누고 txt에 append
#########################################

# 아래내용은 이중 for로 작업해야함.
# 최종결과물은
# txt=['노트북,모니터,키보드,키보드,볼펜,사인펜,노트',
#      '키보드,전자칠판,스케치북,노트북']

pumMok=[['노트북/모니터/키보드','키보드/볼펜','사인펜/노트'],
        ['키보드/전자칠판','스케치북/노트북']]

txt=[]

for pmlst in pumMok:
    for pmtxt in pmlst:
        tmp=""
        for t in pmtxt:
            if t == "/":
                txt.append(tmp)
                tmp=""
            else:
                tmp += t
        if t != "/": 
            txt.append(tmp)

print(txt)

```
