배열 계산

```python
arr1 = np.array([[1,2,3],
                 [6,7,8]])
arr2 = np.array([[10,20,30],
                 [10,20,30]])
```

```python
print(arr1 + arr2)
print(arr2 / arr1)
```

```python
print(arr1.sum())
print(np.sum(arr1))

print(arr1.sum(axis=0))  # 행 방향 합계(열합계)
print(arr1.sum(axis=1))  # 열 방향 합계(행합계)
```

```python
print(arr1.mean())  # 평균
print(arr1.std())  # 표준편차
print(arr1.var())  # 분산
print(arr1.max())  # 최대값
print(arr1.min(axis=0))  # 최소값
```

이미지를 이용한 배열 구조 이해하기
- 고양이 이미지 사용

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


# 이미지 읽어오기
cat_tmp = Image.open('./cat1.jpg')
print(cat_tmp)
plt.imshow(cat_tmp)
```

```python
imgArr=np.array(cat_tmp)
imgArr
```

```python
imgArr.shape
```

고양이 이미지 잘라오기

```python
# 이미지의 상단 잘라오기
cat_top = imgArr[0:280, :, :]
plt.imshow(cat_top)
```

```python
# 이미지의 중앙 오른쪽 출력
cat_rig = imgArr[:, 340:, :]
plt.imshow(cat_rig)
```

```python
# 고양이 눈만 추출하기
plt.imshow(imgArr[200:280, 420:600, :])
```

```python
# 고양이 눈 영역 색상 변경하기
# for문 사용, [200:280, 420:600, 0:3] 포함되면 RGB 값을 모두 50으로 변경

# 행 위치 확인
for i in range(len(imgArr)):
    if i >= 200 and i <= 280:
        for j in range(len(imgArr[i])):  # i번째 행 전체를 가져와 열 위치 확인
            if j >= 420 and j <=600:
                for k in range(3):  # RGB 색상 변경
                    imgArr[i][j][k] = 50

plt.imshow(imgArr)
```
