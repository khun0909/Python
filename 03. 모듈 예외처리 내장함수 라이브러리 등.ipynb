### 예외처리
- try ~ except ~
- 에러 발생시 코드 또는 로직의 문제가 아닌 내부적인 문제로 발생한 에러로 당연히 발생할 수 밖에 없는 에러를 처리하기 위한 방법

```python
# 5개의 데이터를 입력받아 합계를 계산하는 코드
sum=0
for i in range(5):
    try:  # 정상코드
        indata=float(input('숫자를 입력하세요.: '))  # 숫자를 입력 받아 실수형 데이터로 변경
    except: # 에러가 발생했을때 처리 코드
        indata=0
        print('잘못된 숫자 입력으로 현재 값을 0을 처리합니다.')

    sum += indata

print(sum)
```

data 폴더의 pic1, pic2, pic3 파일을 읽어와 파일명, 가로, 세로 형식으로 리스트에 저장

```python
# 1. 3개의 파일을 읽어와 각각의 정보 저장
# 2. imgInfo에 [파일명, 가로, 세로]
# 3. 기본 이미지를 모두 200*200 사이즈로 수정 후 저장(파일명 뒤에 _200 붙여 저장)

from PIL import Image  # 모듈 연결
from tqdm import tqdm  # for문의 진행 상황을 출력

# 디렉토리 경로
fdir='/content/drive/MyDrive/멀캠_프론트엔드_2차/03. Python/코드/'

fname=['pic1', 'pic2', 'pic3'] # 파일 이름
imgInfo=list()  # 파일 정보를 저장할 리스트 객체 생성(imgInfo=[])

for fn in tqdm(fname):
    tmp = Image.open(f'{fdir}data/{fn}.jpg')  # 이미지 읽어오기
    imgInfo.append([fn+'.jpg', tmp.size[0], tmp.size[1]]) #이미지 정보 저장(추가)

    # 이미지 사이즈 조정
    tmpResize=tmp.resize((200, 200))
    # 이미지 저장
    tmpResize.save(f'{fdir}data/{fn}_200.jpg')

print('====== END ======')

```

```python
print(imgInfo)
```

```python
import pandas as pd  # 모듈명이 긴 경우 as를 이용해 별명을 붙일 수 있음

df1 = pd.DataFrame(imgInfo, columns=['filename', 'width', 'height'])
df1
```

```python
# 파일 저장
df1.to_csv('imginfo.csv')
```

### os, glob 모듈을 이용한 폴더의 파일 정보 가져오기
- 개인 구글드라이브의 images 폴더에서 강아지/고양이 이미지를 읽어와 cat과 dog 폴더에 저장

```python
fdir+'images/'
```

```python
# os 모듈 활용
from PIL import Image
import os

fnames=os.listdir(fdir+'images/') # 지정한 폴더에 있는 파일 및 하위폴더 이름을 추출한다.
print(fnames)
```

- 강아지/고양이 파일 읽어와 정보 저장 및 이미지 복사

```python
import tqdm

# 이미지 파일 추출하기
imginfo=[]
for fn in tqdm.tqdm(fnames):
    if fn[-3:] != 'jpg':  # 파일 확장자가 jpg 아니면
        continue
    tmp = Image.open(f'{fdir}images/{fn}')
    imginfo.append({'파일명':fn, '가로':tmp.width, '세로':tmp.size[1]})
    # 강아지/고양이 구분
    if fn[:3] == 'cat':
        tmp.save('./cat/'+fn)
    else:
        tmp.save(f'./dog/{fn}')

print(imginfo)
```
