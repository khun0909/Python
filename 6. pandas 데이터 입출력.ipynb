### open() 함수를 이용한 파일 읽기/쓰기(p175~)
- 파일객체 = open(파일명, 파일열기모드)
- 파일 열기 모드

image.png

```python
# open()을 이용한 텍스트 파일 생성(w)/닫기
f = open('./test.txt', 'w')  # 파일 생성

f.write('파이썬에서 두번째 생성한 텍스트 파일입니다.\n') # 생성파일에 내용 쓰기

f.close() # 파일 닫기
```

```python
# 파일에 내용 추가하기(a)
# with ~ as : f.close() 사용하지 않아도 된다.
with open('./test.txt', 'a') as f:
    for i in range(1, 11):
        data=f'{i}번째 추가된 중입니다.\n' # \n: 줄바꾸기
        f.write(data)
```

- 파일 읽기
    - read() : read(n) 함수는 파일 스트림으로 부터 해당 위치의 문자 n개를 읽어오는 함수, n 생락시 전체를 읽어옴
    - readline(): 파일 스트림으로 부터 해당 위치에서부터 한 줄, 한 라인의 문자열을 읽어옴
    - readlines(): 해당 위치에서부터 파일의 모든 문자열을 읽어옴, 개행을 기준(행)으로 리스트로 저장

```python
# 파일이 있는 폴더 위치 정의
fdir='/content/drive/MyDrive/멀캠_프론트엔드_2차/03. Python/코드/data/'

f = open(fdir + '2023 New Years.txt', 'r')  # txt 파일 읽기 전용으로 연결

readData=f.read(30)  # 파일에서 30개의 글자만 읽어와 출력
readData

f.close()
```

```python
f = open(fdir + '2023 New Years.txt', 'r')

readData=f.read()  # 전체 내용 읽어오기
readData
```

```python
with open(fdir + '2023 New Years.txt', 'r') as f:
    lineData=f.readline()
    print('1번째', lineData)

    lineData=f.readline()
    print('2번째', lineData)
```

```python
with open(fdir + '2023 New Years.txt', 'r') as f:
    linesData=f.readlines()
    print(linesData)
```
